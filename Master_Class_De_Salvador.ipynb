# ELEMENTOS DE TRADING


# 1.- Periodo de Vela

El periodo que se utilice para construir la vela (candlestick) de cotizaciones marca la operativa de un trader. 

No tiene sentido trabajar con velas de 10 segundos para operar en semanas (es un ejemplo tonto que lo explica bastante bien).

Hay que buscar una cierta coherencia entre el periodo de la vela (10", 1', 5', 15'...) y el tiempo durante el que se pretende mantener el trade.

Por ejemplo, el periodo de velas de un minuto funciona muy bien para un horizonte de trade de 1' a 10' (si no da señal de venta, no tendríamos que salir). La vela de 5' funciona muy bien entre 5' y 60' y con esa lógica se pueden seguir construyendo las relaciones.

Velas por debajo del minuto pueden tener sentido, pero los movimientos serán más agresivos. Por lo general, estrategias con velas de mayor duración siempre llevan aparejadas menos operaciones que las velas más cortas. Se puede decir que las velas my cortas se utilizan para scalping (%beneficios muy pequeños), mientras que las velas con mayor periodo van a por movimientos más amplios de precio (parece obvio).

Por otra parte, cuanto más corta sea la vela, menor latencia deberá tener el sistema para poder operar con garantías porque el periodo de "holding" del activo se reduce (nos estaríamos comiendo con la latencia un % muy grande de nuestro tiempo de "holding").

**Para el proyecto que estamos trabajando, un periodo de vela de vela de 1' o 5' puede tener sentido.**

# 2.- Monedas a Integrar

Aunque cada mercado puede tener unas características epecíficas que habría que tener en cuenta (seguro que el de cryptos también las tiene), hay dos que son comunes a todos los mercados: la liquidez y la volatilidad.

1.- Liquidez: es el elemento más importante para hacer trading seguro. En definitiva, es la cantidad total del activo en cuestión (acciones, futuros, opciones, cryptos...) que se ha intercambiado en el mercado en un periodo dado (se suele utilizar el volumen de transacciones diarios como referencia).

Llevado al extremo, si un activo no tiene liquidez y tienes una posición en él, no podrías salir porque no tienes contrapartida.

En masterthecryto.com hay un post que explica cómo medir la liquidez de una crypto:

https://masterthecrypto.com/cryptocurrency-liquidity-measuring-liquidity/


2.- Volatilidad: la volatilidad produce oportunidades de inversión porque indica que el activo se mueve mucho, y mucho movimiento supone cambios de precio importantes y oportunidad de grandes beneficios o grandes pérdidas.

Este elemento es importante, pero puede serlo mucho menos que el primero porque, de hecho, hay muchas estrategias que sólo operan en periodo de poca volatilidad (suele ser la mayor parte del tiempo).

Teniendo en cuenta estos dos factores. El proceso sería, primero, elegir el rango de liquidez que facilita el trading. Una elección sencilla es coger las criptos con mayor liquidez. Esto se puede consultar en "coinmarketcap.com". Se pueden utilizar distintos indicadores de liquidez, pero los resultados van a ser muy parecidos por lo que, si nos quedamos con el Volumen en $ de las últimas 24 horas, el ranking será acertado:

En segundo lugar, habría que escoger entre estas las que tengan mayor volatilidad porque para day tiene sentido operar con estrategias de alta volatilidad. Una buena medida puede ser la variación del precio en los últimos 7 días (también se puede consultar en "coinmarketcap.com"). ## 3.1.- Indicadores Técnicos

Los indicadores técnicos son fórmulas matemáticas basadas en históricos de variables relevantes (precios, volumen...) que se utilizan por los Traders Técnicos para predecir el futuro y tomar decisiones de inversión.

En nuestro caso, podríamos **utilizar los indicadores técnicos populares para construir las variables del modelo y clasificarlos en función de un target fijado.**

## 3.2.- Tipos de indicadores Técnicos

Los Indicadores Técnicos se clasifican de muchas maneras diferentes y es frecuente encontrar indicadores clasificados de distinta forma según el autor que estés siguiendo.

Hay dos grandes aproximaciones a la organización de los indicadores técnicos: leading indicators y lagging indicators.

*Leading Indicators:* son indicadores que intentan anticipar la dirección futura del activo analizado.

*Lagging Indicators:* son indicadores que dan señales una vez que el movimiento ya ha pasdo o está en progreso. Se utilizan para confirmar tendencias antes de entrar en el mercado.

Si combinamos las dos aproximaciones podemos decir que hay 4 grandes tipos de indicadores técnicos:

**1.- Indicadores de Tendencia:** miden la dirección y fuerza de una tendencia, usando algún tipo de media para establecer un benchmark contra el que compararse. Cuando el precio se mueve por encima de la media, se puede interpretar como una señal alcista. Cuando el precio cae por debajo de la media, se cambian las tornas y la señal se convierte en bajista (todo esto tiene matizaciones).

**2.- Indicadores de Momento:** ayudan a identificar la velocidad del movimiento del precio comparando la evolución del precio a lo largo del tiempo (precios de cierre de vela vs precios de cierre de vela anteriores). Se suelen colocar debajo de los gráficos de precio, oscilando cuando los precios cambian.

**3.- Indicadores de volatilidad:** miden el nivel de variación del precio independientemente de la dirección. Es decir, dan una referencia de la dispersión de los retornos a lo largo de un periodo. Bajas volatilidades suelen ir asociadas a movimientos lentos de precio con oscilaciones predecibles de los mismos, mientras que altas volatilidades se asocian a movimientos de precios violentos con oscilaciones poco predecibles. 

**4.- Indicadores de Volumen:** dan una orientación de la fuerza de una tendencia o confirmación de la dirección del mercado. Se asume que los movimientos con volumen están confirmados, mientras los movimientos que no llevan asociado volumen son poco fiables. En ocasiones, el incremento de volumen sin cambio de precio puede ser la antesala de un gran movimiento alcista o bajista.

INDICADORES DE TENDENCIA:

INDICADORES DE MOMENTO

**1.- Oscilador Estocástico (Stochastic Oscillator) - Leading Indicator:** utilizado para predecir puntos de cambio de tendencia al comparar el precio de cierre de la vela contra un rango del precio.

El indicador puede oscilar entre 0 y 100. Por lo general, el áea por encima de 80 indica que te encuentras en una región de sobrecompra. Si el oscilador cruzase a la baja el 80, se podría entender como una señal de venta. Al contrario, el área por debajo de 20 se conoce como área de sobreventa. Si el oscilador cruzase al alza el 20, se podría entender como una señal de compra.

**2.- Relative Strength Index -Leading Indicator:** es un oscilador que se mueve entre 0 y qoo y mide la velocidad y cambio en el movimiento del precio.

A su vez, si el indicador está por debajo de 30, se entiende que el activo está sobrevendido y un cruce al alza de ese nivel podría ser una señal de compra.

INDICADORES DE VOLATILIDAD

**1.- Bandas de Bollinger (Bollinger Bands) - Lagging Indicator:** es un indicador que construye un canal que envuelve el movmiento del precio.

Los límites del canal se construyen sumando (límite superior) y restando (límite inferior) la desviación típica a la media simple del precio.

Las bandas de bollinger utilizan dos parámetros: el periodo y la desviación típica. Los valores por defecto son 20 ara el periodo y 2 para la deviación típica.

cuando el precio está cercano a la banda superior, da señal de venta y, cuando el precio está cercano a la bada inferior, da señal de compra.

**2.- Average True Range (ATR) - Lagging Indicator:** es la media de los rangos verdaderos (se puede ver la explicación gráfica abajo) en periodo especificado.

Generalmente, se utilizan 14 periodos (o velas). Para medir la voltailidad reciente se suelen utilizar periodos de 2 a 10. Para volatilidades a más largo plazo, se suelen utilizar periodos de 20 a 50.

Un ATR que crece indica un incremento en la volatilidad. Una caída en el precio del activo con un incremento en el ATR indicaría un fuerte movimienton bajista.

Este indicador no es direccional y un ATR que aumenta puede señalar presión bajista o alcista si el movimiento es en dirección contraria.

Muy utilizado en trailing stop loss (órdenes que siguen únicamente la dirección del trade con el fin de asegurar beneficios) donde se suelen multiplicar por un factor entre 1 y 3.

INDICADORES DE VOLUMEN

**1.- On Balance Volume (OBV) - Leading Indicator:** este indicador mide la presión compradora y vendedora al sumar el volumen de las velas alcistas y restar el volumen de las velas bajistas.

El valor del indicador OBV no es relevante, lo realmente importante es la dirección del mismo. Cuando tanto el precio como el OBV están en sucesión de máximos y mínimos crecientes, es probable que la tendencia continúe. Cuando tanto el precio como el OBV están en sucesión de máximos y mínimos decrecientes, es probable, también, que la tendencia continúe. 

Si esta correlación en el movimiento se rompe, la tendencia puede dar señales de agotamiento y empezar a fallar.

**2.- Chaikin Money Flow (CMF) -Leading Indicator:** este indicador compara el precio de cierre de la vela con el rango Máximo-Mínimo de la misma vela con el fin de determinar cuánto dinero está entrando o saliendo del activo. Posteriormente aplica este resultado al volumen de la vela para entender el impacto en el volumen. El periodo típico del CMF es 21 periodos (velas).

Un CMF por encima de la línea cero es una señal de fortaleza en el mercado. Por debajo de 0 da señal de debilidad del activo.

CMF lanza una seña de compra cuando el precio alcanza un nuevo mínimo en una zona de sobreventa y el CMF diverge con un mínimo más alto, empezando a crecer.

Calculation
CMF = n-day Sum of [(((C - L) - (H - C)) / (H - L)) x Vol] / n-day Sum of Vol

Where:
n = number of periods, typically 21
H = high
L = low
C = close
Vol = volume

```python
from mpl_finance import candlestick_ohlc, volume_overlay
import talib
import pandas as pd

df = pd.read_csv('https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=MSFT&apikey=demo&datatype=csv',
           index_col=0,
           parse_dates=True,
           infer_datetime_format=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import datetime

import matplotlib.dates as mdates
import matplotlib.pyplot as plt

from matplotlib.dates import MONDAY, DateFormatter, DayLocator, WeekdayLocator
from datetime import timedelta


def plot_pattern(quotes,title,pattern_name,index):
    # quotes : pandas dataframe
    # index : annotation index in 
    # sample : plot_candle2(AP_df,title='Test',pattern_name='Hanging Man',selected_date='2005-04-29',date1='2005-04-07',date2='2005-05-25')        
    mondays = WeekdayLocator(MONDAY)        # major ticks on the mondays
    alldays = DayLocator()              # minor ticks on the days
    weekFormatter = DateFormatter('%b %d')  # e.g., Jan 12
    dayFormatter = DateFormatter('%d')      # e.g., 12
    
    l = quotes.shape[0]
    d1 = index + 16 if index +8 < l else l-1  # 7 day after
    d2 = index - 15 if index - 7 >= 0 else 0 # 7 days before

    # select desired range of dates
    quotes2 = quotes.iloc[range(d2,d1)]
    # print(quotes2.shape)
    
    fig, ax = plt.subplots()
    fig.subplots_adjust(bottom=0.2)
    ax.xaxis.set_major_locator(mondays)
    ax.xaxis.set_minor_locator(alldays)
    ax.xaxis.set_major_formatter(weekFormatter)
    # ax.xaxis.set_minor_formatter(dayFormatter)

    # plot_day_summary(ax, quotes, ticksize=3)
    candlestick_ohlc(ax, zip(mdates.date2num(quotes2.index.to_pydatetime()),
                         quotes2['open'], quotes2['high'],
                         quotes2['low'], quotes2['close']),
                 width=0.6,colordown='red',colorup='green')

    ax.xaxis_date()
    ax.autoscale_view()
    fig.set_size_inches(10,5)
    plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')
    ax.set_title(title)
    ymin, ymax = ax.get_ylim()
    selected_date = quotes.iloc[[index]].index.to_pydatetime()
    
    ax.annotate(pattern_name, 
                xy = (selected_date,quotes.loc[selected_date].high ),
                xytext=(selected_date,quotes.loc[selected_date].high + (ymax-ymin)/4 ),
                arrowprops = dict(facecolor='black',width=0.5,shrink=0.25))
    plt.show()
```

```python
open = df['open'].values
high = df['high'].values
low = df['low'].values
close = df['close'].values
result = talib.CDLHAMMER(open,high,low,close)
```

```python
result
```

```python
hammer_days = np.where(result != 0)[0]
for i in hammer_days :
  plot_pattern(df,'Hammer day '+str(i),'Hammer',i)
```
